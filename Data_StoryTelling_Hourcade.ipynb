Por lo que vamos a aplicar una tecnica para eliminar outliers definiendo una formula aplicando IQR. Cabe destacar que se analizo el metodo Z-Score pero IQR resulto mas efectivo

```python
print('min: ',df.price.min())
print('max: ',df.price.max())
print('mean: ',df.price.mean())
```

```python
fig = plt.figure(figsize=(10, 4))
grid = GridSpec(ncols=2, nrows=1, figure=fig)

# Boxplot
ax1 = fig.add_subplot(grid[:,0])
sns.boxplot(df['price'], ax=ax1)
# Histograma
ax2 = fig.add_subplot(grid[:,1])
sns.histplot(df['price'], ax=ax2, kde=True)
```

```python
#Mediante IQR eliminamos filas con outliers
def remove_outliers(df, x):
    q25 = np.percentile(df[x], 25)
    q75 = np.percentile(df[x], 75)
    iqr = q75 - q25
    cut_off = iqr * 1.5
    lower, upper = 1, (q75 + cut_off)
    df = df[(df[x] < upper) & (df[x] > lower)]
    print(f"Outliers of {x} are removed.\n")
    return df
```

```python
df = remove_outliers(df, 'price')
```

```python
fig = plt.figure(figsize=(10, 4))
grid = GridSpec(ncols=2, nrows=1, figure=fig)

# Boxplot
ax1 = fig.add_subplot(grid[0,0])
sns.boxplot(df['price'], ax=ax1)
# Histograma
ax2 = fig.add_subplot(grid[0,1])
sns.histplot(df['price'], ax=ax2, kde=True)
```

**Precision**
La precisión se define como la proporción de muestras positivas predichas correctamente en relación con todas las muestras que el modelo ha predicho como positivas. Tambien conocida como Tasa de Verdaderos Positivos: casos positivos que fueron correctamente identificada.
En nuestro caso el 99% de todas las muestras predichas como positivas son realmente positivas.

**Recall**
La métrica de "recall" (Sensibilidad) es una medida utilizada en problemas de clasificación para evaluar la capacidad de un modelo para identificar correctamente todas las muestras positivas. También se conoce como "sensibilidad" o "verdadero positivo"
El valor de recall varía entre 0 y 1, donde 1 indica un recall perfecto, es decir, todas las muestras positivas fueron identificadas correctamente, mientras que 0 indica un recall nulo, es decir, ninguna muestra positiva fue identificada correctamente.
En este caso es el 99% de las muestras positivas fueron identificadas correctamente.

**Especificidad**
La especificidad es una métrica utilizada en problemas de clasificación para evaluar la capacidad de un modelo para predecir correctamente las muestras negativas. La especificidad se define como la proporción de muestras negativas predichas correctamente en relación con todas las muestras que el modelo ha predicho como negativas.
El valor de especificidad también varía entre 0 y 1, donde 1 indica una especificidad perfecta, es decir, todas las muestras negativas fueron identificadas correctamente, mientras que 0 indica una especificidad nula, es decir, ninguna muestra negativa fue identificada correctamente.
En nuestro caso el 99% de las muestras negativas fueron identificadas correctamente.


**Accuracy**
Es una métrica comúnmente utilizada para evaluar la calidad de un modelo de clasificación. Se define como la proporción de muestras clasificadas correctamente sobre el total de muestras evaluadas.
En otras palabras, la precisión mide la exactitud general del modelo al predecir tanto las muestras positivas como las negativas.
El valor de precisión varía entre 0 y 1, donde 1 indica una precisión perfecta, es decir, todas las muestras fueron clasificadas correctamente, mientras que 0 indica una precisión nula, es decir, ninguna muestra fue clasificada correctamente.
En nuestro caso el 99% de las muestras fueron clasificadas correctamente.

**F1 Score**
Esta es otra métrica muy empleada porque nos resume la (Precisión) Precisión y Sensibilidad (Recall) en una sola métrica.
Un valor alto del 0.99 como el obtenido indica pocos Falsos Positivos y pocos Falsos Negativos, identificando que las amenazas reales son muy bajas.
